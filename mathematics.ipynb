```python
dataset=pd.read_csv("Employee_dataset.csv")
```

```python
dataset.head()
```

```python
dataset.sample(5)
```

```python
mn=np.mean(dataset["age"])
```

```python
dataset["age"].mean()
```

```python
sns.histplot(x='age',data=dataset,bins=[i for i in range(0,60,10)])
plt.plot([mn for i in range(0,12000)],[i for i in range(0,12000)],c='green')
plt.show()
```

### b.median

```python
dataset.isnull().sum()
```

```python
np.median(dataset['age'])
```

```python
nm=dataset['age'].median()
```

```python
nm
```

```python

sns.histplot(x='age',data=dataset,bins=[i for i in range(0,60,10)])
plt.plot([mn for i in range(0,12000)],[i for i in range(0,12000)],c='green')
plt.plot([nm for i in range(0,8000)],[i for i in range(0,8000)],c='blue')
plt.show()
```
