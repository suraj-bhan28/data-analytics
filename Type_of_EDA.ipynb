# Exploratory Data Analysis in Python | Set 1

```python
#Step 1: Setting Up Environment
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 

from scipy.stats import trim_mean
```

```python
#Step 2: Loading and Inspecting the Dataset
data = pd.read_csv("state.csv") 

print ("Type : ", type(data), "\n\n") 

print ("Head -- \n", data.head(10)) 

print ("\n Tail -- \n", data.tail(10))
```

```python
#Step 3: Adding and Modifying Columns
data['populationInMillion']=data['Population']/1000000
data.sample(5)
```

```python
# renaming column
data.rename(columns={'Murder.Rate':'Murder_rate'},inplace=True)
list(data)

# data.columns
```

```python
#Step 4: Describing the Data
data.describe()
```

```python
data.info()
```

```python
#Step 5: Calculating Central Tendencies
#1. Mean
Population_mean = data.Population.mean() 
print ("Population Mean : ", Population_mean) 
  
MurderRate_mean = data.Murder_rate.mean() 
print ("\nMurderRate Mean : ", MurderRate_mean)
```

```python
#2. Trimmed Mean
#Trimmed mean calculates the average by removing a certain percentage of the highest and 
#lowest values in the dataset. This helps reduce the impact of outliers or extreme values that could skew the overall mean
population_TM = trim_mean(data.Population, 0.1) 
print ("Population trimmed mean: ", population_TM) 
  
murder_TM = trim_mean(data.Murder_rate, 0.1) 
print ("\nMurderRate trimmed mean: ", murder_TM)
```

```python
#3. Weighted Mean
murderRate_WM = np.average(data.Murder_rate, weights = data.Population) 
print ("Weighted MurderRate Mean: ", murderRate_WM)
```

```python
#4. Median
Population_median = data.Population.median() 
print ("Population median : ", Population_median) 
  
MurderRate_median = data.Murder_rate.median() 
print ("\nMurderRate median : ", MurderRate_median)
```

```python
# visusalization
# Plot Population In Millions
fig, ax1 = plt.subplots()
fig.set_size_inches(15,  9)

ax1 = sns.barplot(x ="State", y ="Population", 
                  data = data.sort_values('Murder_rate'), 
                                        palette ="Set2")
ax1.set(xlabel ='States', ylabel ='Population In Millions')
ax1.set_title('Population in Millions by State', size = 20)

plt.xticks(rotation =-90)
```

```python
# Plot Murder Rate per 1, 00, 000

fig, ax2 = plt.subplots()
fig.set_size_inches(15,  9)


ax2 = sns.barplot(
    x ="State", y ="MurderRate", 
    data = data.sort_values('MurderRate', ascending = 1), 
                                         palette ="husl")

ax2.set(xlabel ='States', ylabel ='Murder Rate per 100000')
ax2.set_title('Murder Rate by State', size = 20)

plt.xticks(rotation =-90)
```

```python
Population_std = data.Population.std()
print ("Population std : ", Population_std)

MurderRate_std = data.Murder_rate.std()
print ("\nMurderRate std : ", MurderRate_std)
```

```python
Population_var = data.Population.var()
print ("Population var : ", Population_var)

MurderRate_var = data.Murder_rate.var()
print ("\nMurderRate var : ", MurderRate_var)
```

```python
# Inter Quartile Range of Population
population_IQR = data.Population.describe()['75 %'] - data.Population.describe()['25 %']

print ("Population IQR : ", population_IRQ)

# Inter Quartile Range of Murder Rate
MurderRate_IQR = data.Murder_rate.describe()['75 %'] - data.Murder_rate.describe()['25 %']

print ("\nMurderRate IQR : ", MurderRate_IQR)
```
